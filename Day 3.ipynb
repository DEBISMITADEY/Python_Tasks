1) What are Data Structures? List some of data structures and their use in real world application? <br><br>
Data structure rae the fundamental components in computer science which enable efficient organisation, storage, and manipulation of data within a computer system. They establish a systematic approach to managing data, facilitating streamlined access, insertion, deletion, and search operations. Diverse data structures have been devised to address varying data types and operational requirements.<br>
Some of the data structures and their real world applications are:<br>
(a) Arrays: Arrays consist of a collection of elements stored in contiguous memory locations. They are particularly useful for storing and accessing fixed-size sequences of elements. Arrays find extensive application in scenarios necessitating the storage of item lists, matrix representations, and the implementation of other data structures.<br>
(b) Linked Lists: Linked lists are linear data structures wherein each element (referred to as a node) comprises a value and a reference to the subsequent node. Linked lists prove invaluable when frequent insertions and deletions are required. They are often employed in the implementation of stacks, queues, and other dynamic data structures.<br>
(c) Stacks: Stacks adhere to the Last-In-First-Out (LIFO) principle, wherein elements are exclusively added to and removed from one end, commonly known as the top. Recursive function calls, expression evaluation, and backtracking algorithms are among the many domains that extensively utilize stacks.<br>
(d) Queues: Queues abide by the First-In-First-Out (FIFO) principle, where elements are added at the rear and removed from the front. They find application in situations such as job scheduling, request handling, and breadth-first search algorithms.<br>
(e) Trees: Trees represent hierarchical data structures composed of nodes interconnected by edges. They serve diverse purposes, such as representing hierarchical relationships (e.g., file systems), optimizing data retrieval (e.g., binary search trees), and implementing decision trees for algorithmic processes. <br>
(f) Graphs: Graphs are non-linear data structures comprising vertices (nodes) and edges connecting them. Social networks, transportation networks, and web page ranking algorithms rely heavily on graphs to model and analyze complex relationships.<br>
(g) Hash Tables: Hash tables, or hash maps, store data in key-value pairs. Utilizing hash functions, they map keys to their corresponding values, facilitating efficient lookup, insertion, and deletion operations. Dictionaries, caches, and database indexing heavily employ hash tables.<br>
(h) Heaps: Heaps are specialized trees designed to retrieve the maximum or minimum element efficiently. They serve crucial roles in priority queues, scheduling algorithms, and graph algorithms such as Dijkstra's algorithm.<br><BR>

4) What is difference between ordered data type and unordered data type? <br>
An ordered data type maintains a specific order of its elements. This means that the elements have a defined sequence or arrangement that is consistent and predictable. Elements in an ordered data type can be accessed and retrieved based on their position or index within the data structure. Examples of ordered data types include lists, arrays, and tuples in Python. Whereas An unordered data type does not have a specific order or sequence of its elements. The elements are stored in a manner that does not guarantee a particular arrangement or position. In other words, the order of elements in an unordered data type is arbitrary and may vary each time the data structure is accessed. Examples of unordered data types include sets and dictionaries in Python.
