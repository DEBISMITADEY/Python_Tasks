4) Write Down difference between interpreter and compiler. <br><br>
An interpreter reads and executes the code line by line, translating and executing each statement in real-time. Whereas A compiler translates the entire source code into machine-readable instructions, known as object code or bytecode. <br>
An interpreter directly interprets the source code without the need for a separate compilation step. Whereas a compiler performs a full analysis of the code and generates optimized executable files or libraries. <br>
A compiler and interpreter varies in speed. Compiled code generally runs faster since it has been optimized during the compilation process. Interpreted code tends to be slower as it is translated and executed in real-time. <br>
They have diff in error detection.
Interpreters typically detect errors on-the-fly, stopping execution when an error is encountered. Compilers detect errors during the compilation process and produce error messages before generating the executable code.<br>
An interpreter executes code directly, line by line, while a compiler translates the entire code into executable form.
